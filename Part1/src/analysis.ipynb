```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

from helper import plot
import os
from Part1.resources.constants import RESOURCES_DIRECTORY

times_df = pd.read_csv(os.path.join(RESOURCES_DIRECTORY,"data/times.csv"), index_col=False)
print(times_df.style.to_latex())
```

```python
time_sum = 0
for i in range(1, times_df.shape[1], 1):
    time_sum += sum(times_df.iloc[:,i])

print(f"Tempo di analisi totale: {time_sum} secondi\n{time_sum/60} minuti\n{time_sum/3600} ore")
```
